# Probelma 1 -  Raising Bacteria




#### Prompt ingresado en chatGPT:



#### Solucion:

```python
x = int(input())
count = 0

while x > 0:
    if x & 1:
        count += 1
    x >>= 1

print(count)
```

Este problema chatGPT logró resolverlo correctamente.

# Probelma 2 -  Distinct Numbers



#### Prompt ingresado en chatGPT:



#### Solucion:

```python
n = int(input())
numbers = list(map(int, input().split()))

distinct_numbers = set(numbers)
distinct_count = len(distinct_numbers)

print(distinct_count)
```
